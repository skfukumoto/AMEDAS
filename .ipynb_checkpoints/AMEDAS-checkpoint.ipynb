```python
import pandas as pd

# データセットの読み込み
# AMEDAS_df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/AMEDAS_data3.csv', header=4, usecols=[0,1,4,7,10], encoding='shift_jis', index_col=0 ) 
AMEDAS_df = pd.read_csv('/home/pi/jupyter/AMEDAS/AMEDAS_data3.csv', header=4, usecols=[0,1,4,7,10], encoding='shift_jis', index_col=0 ) 
# 列名を付与
AMEDAS_df.columns = ['Sapporo', 'Naha', 'South Pole', 'Kobe'] 

AMEDAS_df.index = pd.to_datetime( AMEDAS_df.index )

AMEDAS_df.head(20) # 最初の20列データ
#AMEDAS_df.mean() # 平均値
#AMEDAS_df.var() # 分散
#AMEDAS_df.count() # データ数
#AMEDAS_df.describe() # 統計レビュー
#AMEDAS_df.corr() # 相関値
```

```python
import matplotlib.pyplot as plt
plt.scatter(x=AMEDAS_df['Kobe'], y=AMEDAS_df['Sapporo']) # 散布図をプロット

plt.title('Scatter Plot of Kobe vs Sapporo')             # 図のタイトル
plt.xlabel('Kobe') # x軸のラベル
plt.ylabel('Sapporo') # y軸のラベル
plt.grid(True)                                         # グリッド線を表示

plt.show()                                             # 図の表示
```

```python
AMEDAS1_df = AMEDAS_df.sort_values(by='Sapporo', ascending=True)
AMEDAS1_df.head(10)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
#sns.distplot(sensor_df['Sensor1'])
sns.distplot(AMEDAS_df)
sns.pairplot(AMEDAS_df)
```

```python
import matplotlib.pyplot as plt
plt.plot(AMEDAS_df.index, AMEDAS_df['Sapporo']) # 折れ線グラフをプロット

plt.title('Sapporo')          # 図のタイトル
plt.xlabel('Year-Month')                          # x軸のラベル
plt.ylabel('sapporo')                               # y軸のラベル

plt.xticks(rotation=90)

plt.show()                                        # 図の表示
```

```python
import matplotlib.pyplot as plt
# 複数の折れ線グラフをプロット：線の色（color）・点のマーカー（marker）・線の種類（linetype）・凡例表示のためのラベル（label）でグラフを区別
plt.plot(AMEDAS_df.index, AMEDAS_df['Kobe'], color = 'red', marker = 'o', linestyle = '-', label = 'Kobe')
plt.plot(AMEDAS_df.index, AMEDAS_df['Sapporo'], color = 'green', marker = 'x', linestyle = '--', label = 'Sapporo')
plt.plot(AMEDAS_df.index, AMEDAS_df['South Pole'], color = 'blue', marker = 's', linestyle = ':', label = 'South Pole')
plt.plot(AMEDAS_df.index, AMEDAS_df['Naha'], color = 'orange', marker = 'v', linestyle = '-.', label = 'Naha')

plt.title('Transition of Temperarture') # 図のタイトル
plt.xlabel('Year/Month')                  # x軸のラベル

plt.xticks(rotation=90)

plt.ylabel('Temperature (C)')                     # y軸のラベル
plt.legend(loc = 'upper left')         # 凡例を右上（upper right）に表示
#plt.xlim([50, 51])                      # x軸方向の表示範囲を指定
plt.ylim([-30, 35])                    # y軸方向の表示範囲を指定

plt.show()                              # 図の表示
```

```python
import matplotlib.pyplot as plt
fig, axe = plt.subplots(2, 2) # 図を2行2列に分割

# 各領域で折れ線グラフをプロット
axe[0, 0].plot(AMEDAS_df.index, AMEDAS_df['Kobe'], color = 'red', marker = 'o', linestyle = '-', label = 'Kobe')
axe[0, 1].plot(AMEDAS_df.index, AMEDAS_df['Sapporo'], color = 'green', marker = 'x', linestyle = '--', label = 'Sapporo')
axe[1, 0].plot(AMEDAS_df.index, AMEDAS_df['South Pole'], color = 'blue', marker = 's', linestyle = ':', label = 'South Pole')
axe[1, 1].plot(AMEDAS_df.index, AMEDAS_df['Naha'], color = 'orange', marker = 'v', linestyle = '-.', label = 'Naha')

# 各グラフのタイトル・x軸のラベル・y軸のラベルを設定
axe[0, 0].set_title('Kobe')
axe[0, 0].set_xlabel('Year/Month')
axe[0, 0].set_ylabel('Temp (C)')
axe[0 ,0].set_ylim(0, 40)

axe[0, 1].set_title('Sapporo')
axe[0, 1].set_xlabel('Year/Month')
axe[0, 1].set_ylabel('Temp (C)')
axe[0 ,1].set_ylim(-10, 30)

axe[1, 0].set_title('South Pole')
axe[1, 0].set_xlabel('Year/Month')
axe[1, 0].set_ylabel('Temp (C)')
axe[1 ,0].set_ylim(-30, 10)

axe[1, 1].set_title('Naha')
axe[1, 1].set_xlabel('Year/Month')
axe[1, 1].set_ylabel('Temp (C)')
axe[1, 1].set_ylim(0, 40)

plt.tight_layout() # ラベル・タイトルが被らないように配置
plt.show()         # 図の表示
```
