```python
import matplotlib.pyplot as plt
import numpy as np

n = 15
a = (0,0)
b = (0,1)
points = [a,b]
theta = np.pi/2
l = 1
p = 0.7
angle = np.pi/20

def tree(a,b,theta,l,n):
	if n == 0:
		return
	s = (b[0]+l*np.cos(theta + angle),b[1]+l*np.sin(theta + angle))
	t = (b[0]+l*np.cos(theta - angle),b[1]+l*np.sin(theta - angle))
	points.append(s)
	tree(b,s,theta + angle,l*p,n-1)
	points.append(b)
	points.append(t)
	points.append(b)
	points.append(t)
	tree(b,t,theta - angle,l*p,n-1)
	points.append(b)


tree(a,b,theta,l,n)
x = [points[i][0] for i in range(len(points))]
y = [points[i][1] for i in range(len(points))]
fig,ax = plt.subplots()
plt.plot(x,y)
ax.set_xlim([-4,4])
ax.set_ylim([0,6])
plt.show()
```
